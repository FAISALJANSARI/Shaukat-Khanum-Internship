```python
import google.generativeai as genai
from google.colab import userdata
# adding api key

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# geting filtered data from file
file_path='/content/drive/MyDrive/Colab Notebooks/Testing Model (Tokanization issue)/filterd_file.txt'
with open(file_path, 'r', encoding='utf-8') as file:
  # Read the contents of the file and concatenate
  file_contents = file.read()

file_contents
```

```python
# Set up the model

generation_config = {
  "temperature": 0.8,
  "top_p": 1,
  "top_k": 1,
  "max_output_tokens": 2048,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
]

```

```python
# creating moddel instance
model = genai.GenerativeModel(model_name="gemini-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)
```

**Input Prompt**

"Please provide a summary for the following text and do not add any EXTRA TEXT ONLY PROVIDE SUMMARY an do not provide input text in output. Please use the provided symptoms, procedure,chemo and radiation details(if any), tests and results in the summary and please do not add patients personal information in adition generate, summary in a way a doctor would, for the patient to take it to an other hospital  : {file_contents}"


```python
# context input text
prompt_parts = [
    f"Please provide a summary for the following text and do not add any EXTRA TEXT ONLY PROVIDE SUMMARY an do not provide input text in output. Please use the provided symptoms, procedure,chemo and radiation details(if any), tests and results in the summary and please do not add patients personal information in adition generate, summary in a way a doctor would, for the patient to take it to an other hospital  : {file_contents}",
]
```

```python
# genrating summary
esponse = model.generate_content(prompt_parts)
```
